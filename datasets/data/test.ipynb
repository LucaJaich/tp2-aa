```python
import pandas as pd
```

```python
df_films = pd.read_csv('film_reviews.csv', sep="|")
df_films.head()
```

```python
reviews = df_films['review_text'].tolist()
# with regex, only keep letters, letters with "acentos" , spaces, commas, periods and question marks
import re
reviews = [re.sub(r'[^a-zA-ZÀ-ÿ\s,?.]', '', review) for review in reviews]
# remove extra spaces
reviews = [re.sub(r'\s+', ' ', review).strip() for review in reviews]
reviews
```

```python
import json
with open('film_reviews.json', 'w+') as f:
    json.dump(reviews, f, ensure_ascii=False, indent=4)
```
